```python
def self_attention(K,Q,V):
  d_k = Q.shape[-1]
  # to get the embedding dimension shape
  scores = torch.matmul(Q,K.T)/torch.sqrt(torch.tensor(d_k,dtype=torch.float))
  weights = torch.softmax(scores,dim=-1)
  return torch.matmul(weights,V)
```

```python
Q = torch.tensor([[1.0,0.0],[0.0,1.0],[1.0,1.0]])
```

```python
K = Q
```

```python
V = Q
```

```python
self_attention(K,Q,V)
```

```python
import math
def positional_encoding(pos, d):
  pe = []
  for i in range(d):
     if i % 2 == 0:
      pe.append(math.sin(pos / (10000 ** (i/d))))
     else:
      pe.append(math.cos(pos / (10000 ** ((i-1)/d))))
  return torch.tensor(pe)
print(positional_encoding(1, 6))
```
